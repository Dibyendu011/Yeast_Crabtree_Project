```python
import numpy as np
import pandas as pd
import os
from pathlib import Path
import random
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

positive_seq = list(os.listdir("crabtree_positive"))
negative_seq = list(os.listdir("crabtree_negative"))
len(positive_seq), len(negative_seq)
```

```python
random.shuffle(positive_seq)
positive_train = positive_seq[0:round(0.75*len(positive_seq))]
positive_test = positive_seq[0:round(0.25*len(positive_seq))]
len(positive_train), len(positive_test)
```

```python
random.shuffle(negative_seq)
negative_train = negative_seq[0:round(0.75*len(negative_seq))]
negative_test = negative_seq[0:round(0.25*len(negative_seq))]
len(negative_train), len(negative_test)
```

```python
for files in positive_train :
    with open("crabtree_positive"/Path(files), "r") as src, open("train.fasta", 'a') as dest:
        dest.write(src.read()+"\n")

for files in negative_train :
    with open("crabtree_negative"/Path(files), "r") as src, open("train.fasta", 'a') as dest:
        dest.write(src.read()+"\n")
```

```python
seq = []
from Bio import SeqIO
for sequence in SeqIO.parse("train.fasta","fasta"):
    seq.append(str(sequence.seq))
train_df = pd.DataFrame(seq, columns=["sequence"])
sense_list = ["positive"]*len(positive_train) + ["negative"]*len(negative_train)
train_df["sense"]=sense_list
train_df = train_df.sample(frac=1).reset_index(drop=True)
train_df
```

```python
for files in positive_test :
    with open("crabtree_positive"/Path(files), "r") as src, open("test.fasta", 'a') as dest:
        dest.write(src.read()+"\n")

for files in negative_test :
    with open("crabtree_negative"/Path(files), "r") as src, open("test.fasta", 'a') as dest:
        dest.write(src.read()+"\n")
```

```python
seq2 = []
from Bio import SeqIO
for sequence in SeqIO.parse("test.fasta","fasta"):
    seq2.append(str(sequence.seq))
test_df = pd.DataFrame(seq2, columns=["sequence"])
sense_list2 = ["positive"]*len(positive_test) + ["negative"]*len(negative_test)
test_df["sense"]=sense_list2
test_df = test_df.sample(frac=1).reset_index(drop=True)
test_df
```

```python
def Kmers_funct(seq, size=6):
    return [seq[x:x+size].lower() for x in range(len(seq) - size + 1)]

train_df['words'] = train_df.apply(lambda x: Kmers_funct(x["sequence"]), axis=1)
train_df = train_df.drop('sequence', axis=1)
train_label = np.array(train_df["sense"])

test_df['words'] = test_df.apply(lambda x: Kmers_funct(x["sequence"]), axis=1)
test_df = test_df.drop('sequence', axis=1)
test_label = np.array(test_df["sense"])
```

```python
train_texts = list(train_df["words"])
for item in range(len(train_texts)):
    train_texts[item] = " ".join(train_texts[item])

test_texts = list(test_df["words"])
for item in range(len(test_texts)):
    test_texts[item] = " ".join(test_texts[item])
```

```python
cvec = CountVectorizer(ngram_range=(4,4))
train_vec = cvec.fit_transform(train_texts)
test_vec = cvec.transform(test_texts)
print(train_vec.shape)
print(test_vec.shape)
```

```python
model = MultinomialNB(alpha=0.1)
model.fit(train_vec, train_label)
sense_pred = model.predict(test_vec)
```

```python
table = pd.DataFrame(sense_pred, columns=["Predicted Labels"])
table["True Labels"] = test_label
table
```

```python
print(classification_report(test_label,sense_pred))
```
